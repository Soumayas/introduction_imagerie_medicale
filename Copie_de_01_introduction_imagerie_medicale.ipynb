9) Comparez ce que vous avez obtenu pour le fichier .jpg ainsi que le fichier .dcm et expliquez.

```python
import cv2
import pydicom
import numpy as np

# Chemins des fichiers
chemin_jpg = 'Copie de brain_067.jpg'
chemin_dicom = 'brain_067.dcm'

# Fonction pour obtenir la valeur maximale des niveaux de gris
def valeur_maximale_niveaux_gris(chemin):
    if chemin.endswith('.jpg'):
        image = cv2.imread(chemin, cv2.IMREAD_GRAYSCALE)
        return np.max(image)
    elif chemin.endswith('.dcm'):
        image_dicom = pydicom.dcmread(chemin)
        return np.max(image_dicom.pixel_array)
    else:
        raise ValueError("Format de fichier non pris en charge")

# Comparaison des valeurs maximales
valeur_max_jpg = valeur_maximale_niveaux_gris(chemin_jpg)
valeur_max_dicom = valeur_maximale_niveaux_gris(chemin_dicom)

# Affichage des résultats
print(f"Valeur maximale des niveaux de gris pour le fichier .jpg : {valeur_max_jpg}")
print(f"Valeur maximale des niveaux de gris pour le fichier .dcm : {valeur_max_dicom}")
```

Explication : La fonction valeur_maximale_niveaux_gris prend en paramètre le chemin du fichier, lit l'image (pour le fichier .jpg) ou le fichier DICOM, puis retourne la valeur maximale des niveaux de gris.

Les valeurs maximales des niveaux de gris pour les fichiers .jpg et .dcm sont obtenues à l'aide de cette fonction.

Les résultats sont ensuite affichés.

4) Les données DICOM se présentent généralement sous forme d'un ensemble de coupes 2D constituant un volume global 3D. Le répertoire DICOM_3D_volume contient un exemple d'ensemble de coupes 2D d'un cerveau humain. Stockez les noms de tous les fichiers contenus dans le répertoire DICOM_3D_volume dans une liste que vous appellerez list_filenames.

```python
import os

# Chemin du répertoire DICOM_3D_volume
chemin_repertoire_dicom = '/content/drive/MyDrive/LAB TICV/DICOM_3D_volume'

# Initialiser une liste pour stocker les noms de fichiers
list_filenames = []

try:
    # Parcourir les fichiers du répertoire
    for filename in os.listdir(chemin_repertoire_dicom):
        # Vérifier si le chemin est un fichier
        if os.path.isfile(os.path.join(chemin_repertoire_dicom, filename)):
            # Ajouter le nom du fichier à la liste
            list_filenames.append(filename)

    # Afficher la liste des noms de fichiers
    print("Liste des noms de fichiers dans le répertoire DICOM_3D_volume:")
    for filename in list_filenames:
        print(filename)

except FileNotFoundError:
    print(f"Répertoire '{chemin_repertoire_dicom}' non trouvé. Assurez-vous que le répertoire est dans le bon emplacement.")
```

5) En utilisant les lignes de code suivantes, créez un volume de données 3D regroupant toutes les coupes.

```python
import os
import pydicom
import numpy

# Chemin du répertoire DICOM_3D_volume
dicom_volume_path = '/content/drive/MyDrive/LAB TICV/DICOM_3D_volume'
# Get a reference file
reference_dataset = pydicom.dcmread(os.path.join(dicom_volume_path, list_filenames[0]))

# Load dimensions based on the number of rows, columns, and slices (along the Z axis)
ConstPixelDims = (int(reference_dataset.Rows), int(reference_dataset.Columns), len(list_filenames))

# Load spacing values (in mm)
ConstPixelSpacing = (float(reference_dataset.PixelSpacing[0]), \
                     float(reference_dataset.PixelSpacing[1]), \
                     float(reference_dataset.SliceThickness))

print("volume dimensions : ", ConstPixelDims)
print("volume pixel spacing : ", ConstPixelSpacing)


# The array is sized based on 'ConstPixelDims'
ArrayDicom = numpy.zeros(ConstPixelDims, dtype=reference_dataset.pixel_array.dtype)

# loop through all the DICOM files
for filenameDCM in list_filenames:
    # read the file
    ds = pydicom.dcmread(os.path.join(dicom_volume_path, filenameDCM))
    # store the raw image data
    ArrayDicom[:, :, list_filenames.index(filenameDCM)] = ds.pixel_array
```

6) Ecrivez la fonction explore_3dimage_axial qui permet d'afficher le volume 3D en parcourant les coupes selon la direction z (vue axiale).

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import ipywidgets as widgets
from ipywidgets import interact, fixed

def explore_3dimage_axial(volume):
    """
    Display axial slices of a 3D volume.

    Parameters:
    - volume: 3D NumPy array representing the volume.
    """
    num_slices = volume.shape[2]

    def update_slice(z):
        plt.imshow(volume[:, :, z], cmap='gray')
        plt.title(f"Axial Slice: {z}")
        plt.show()

    # Create an interactive slider
    interact(update_slice, z=widgets.IntSlider(min=0, max=num_slices-1, step=1, value=0))

# Assuming you have a 3D volume stored in the 'ArrayDicom' variable
explore_3dimage_axial(ArrayDicom)
```

7) Ecrivez les fonctions explore_3dimage_coronal et explore_3dimage_sagittal qui permettent de représenter le volume 3D respectiviement en vue coronale et sagittale et appliquez les au volume de données précédent.

```python
import ipywidgets as widgets
from ipywidgets import interact, fixed

def explore_3dimage_coronal(volume):
    """
    Display coronal slices of a 3D volume.

    Parameters:
    - volume: 3D NumPy array representing the volume.
    """
    num_slices = volume.shape[0]

    def update_slice(y):
        plt.imshow(volume[y, :, :], cmap='gray', vmin=0, vmax=np.max(volume))
        plt.title(f"Coronal Slice: {y}")
        plt.show()

    # Create an interactive slider
    interact(update_slice, y=widgets.IntSlider(min=0, max=num_slices-1, step=1, value=0))

def explore_3dimage_sagittal(volume):
    """
    Display sagittal slices of a 3D volume.

    Parameters:
    - volume: 3D NumPy array representing the volume.
    """
    num_slices = volume.shape[1]

    def update_slice(x):
        plt.imshow(volume[:, x, :], cmap='gray', vmin=0, vmax=np.max(volume))
        plt.title(f"Sagittal Slice: {x}")
        plt.show()

    # Create an interactive slider
    interact(update_slice, x=widgets.IntSlider(min=0, max=num_slices-1, step=1, value=0))



# Display coronal slices
explore_3dimage_coronal(ArrayDicom)
# Display sagittal slices
explore_3dimage_sagittal(ArrayDicom)
```
